```python
#Q8-Write a class named arithmetic with method for calculating addition, substraction, multiplication and division of two numbers.
class arithmetic:

    def addition(self, x, y):
        return x + y

    def substraction(self, x, y):
        return x - y

    def multiplication(self, x, y):
        return x * y

    def divide(self, a, b):
        if b == 0:
            return "Error: Division by zero is not allowed."
        else:
            return a / b


calc = arithmetic()

print(f"Addition: {calc.addition(3, 4)}")
print(f"Subtraction: {calc.substraction(3, 4)}")
print(f"Multiplication: {calc.multiplication(3, 4)}")
print(f"Division: {calc.divide(3, 4)}")
```

```python
#Q9- write a for loop or method  to calculate summation of first n numbers
#i.e if n=3, it will print or return 6 (1+2+3)
def calculate_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

n = 3
print(f"The sum of the first {n} numbers is: {calculate_sum(n)}")
```

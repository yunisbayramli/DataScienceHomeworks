# Case Intro
Term deposits are a major source of income for a bank. A term deposit is a cash investment held at a financial institution. Your money is invested for an agreed rate of interest over a fixed amount of time, or term. The bank has various outreach plans to sell term deposits to their customers such as email marketing, advertisements, telephonic marketing, and digital marketing.

Telephonic marketing campaigns still remain one of the most effective way to reach out to people. However, they require huge investment as large call centers are hired to actually execute these campaigns. Hence, it is crucial to identify the customers most likely to convert beforehand so that they can be specifically targeted via call.

The data is related to direct marketing campaigns (phone calls) of a Portuguese banking institution. The classification goal is to predict if the client will subscribe to a term deposit (variable y).

Content
The data is related to the direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or not ('no') subscribed by the customer or not. The data folder contains two datasets:-

Bank.csv: 45,211 rows and 18 columns ordered by date (from May 2008 to November 2010)

Detailed Column Descriptions
bank client data:

1 - age (numeric)

2 - job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student",
"blue-collar","self-employed","retired","technician","services")

3 - marital : marital status (categorical: "married","divorced","single"; note: "divorced" means divorced or widowed)

4 - education (categorical: "unknown","secondary","primary","tertiary")

5 - default: has credit in default? (binary: "yes","no")

6 - balance: average yearly balance, in euros (numeric)

7 - housing: has housing loan? (binary: "yes","no")

8 - loan: has personal loan? (binary: "yes","no")
# related with the last contact of the current campaign:
9 - contact: contact communication type (categorical: "unknown","telephone","cellular")
10 - day: last contact day of the month (numeric)

11 - month: last contact month of year (categorical: "jan", "feb", "mar", …, "nov", "dec")

12 - duration: last contact duration, in seconds (numeric)

# other attributes:
13 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)

14 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)

15 - previous: number of contacts performed before this campaign and for this client (numeric)

16 - poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")

Output variable (desired target):

17 - y - has the client subscribed a term deposit? (binary: "yes","no")

Missing Attribute Values: None

```python
# Following function converts non-numeric variables (e.g., 'category', 'object') into numeric using label encoding
# Note:
#Label encoding converts categorical values into integer codes. Each unique category is assigned a distinct number, such as 1, 2, 3, 4, etc.
#Example : In our data, education variable has following values:{ 'primary' 'secondary', 'tertiary', 'unknown'}
#and it is mapped { 'primary': 0, 'secondary': 1, 'tertiary': 2, 'unknown': 3}

#  education  education_encoded
#  secondary                  1
#   tertiary                  2
#    primary                  0
#    unknown                  3
#   tertiary                  2
#    primary                  0
#
#Label encoding is generally suitable for tree-based models (e.g., decision trees, random forests, boosting methods).
#However, it may not be appropriate for models where the objective function relies on distance-based calculations, such as neural networks, support vector machines (SVM), or linear regression.
#For nominal features (categories with no intrinsic ordering, e.g., "red," "blue," "green"), label encoding can mislead the model by implying an ordinal relationship where none exists.
#In such cases, one-hot encoding is usually preferred.
#One-hot encoding creates binary columns (dummy variables) for each category in a categorical feature, avoiding the introduction of unintended ordinality.

def Encoder(df):
          from sklearn import preprocessing
          columnsToEncode = list(df.select_dtypes(include=['category','object']))
          le = preprocessing.LabelEncoder()
          for feature in columnsToEncode:
              try:
                  df[feature] = le.fit_transform(df[feature])
              except:
                  print('Error encoding '+feature)
          return df
```

```python
#QUESTION 1

# ==============================
# 1. Import Libraries
# ==============================
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Install if needed:
# !pip install xgboost lightgbm

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier


# ==============================
# 2. Load Data
# ==============================
df = pd.read_csv(
    'https://raw.githubusercontent.com/ogut77/DataScience/main/data/Bank.csv',
    sep=';'
)


# ==============================
# 3. Encode Categorical Variables
# ==============================
from sklearn.preprocessing import LabelEncoder

def Encoder(df):
    df = df.copy()
    cols = df.select_dtypes(include=['object']).columns
    le = LabelEncoder()

    for col in cols:
        df[col] = le.fit_transform(df[col])

    return df

df = Encoder(df)


# ==============================
# 4. Split Features & Target
# ==============================
y = df['y']   # already encoded to 0/1
X = df.drop('y', axis=1)


# ==============================
# 5. Train-Test Split
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=17
)


# ==============================
# 6. Initialize Models
# ==============================
models = {
    "Random Forest": RandomForestClassifier(random_state=17),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=17),
    "LightGBM": LGBMClassifier(random_state=17),
    "Gradient Boosting": GradientBoostingClassifier(random_state=17)
}


# ==============================
# 7. Train & Evaluate Models
# ==============================
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

    print(f"{name} Accuracy: {acc:.4f}")


# ==============================
# 8. Find Best Model
# ==============================
best_model = max(results, key=results.get)

print("\nBest Model:", best_model)
print("Best Accuracy:", results[best_model])
```

```python
#QUESTION 2

# =========================================
# QUESTION 2
# Optuna hyperparameter tuning (100 trials)
# =========================================

# Install if needed:
!pip install optuna xgboost lightgbm

import pandas as pd
import numpy as np
import optuna

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier


# =========================================
# 1. Load Data
# =========================================
df = pd.read_csv(
    'https://raw.githubusercontent.com/ogut77/DataScience/main/data/Bank.csv',
    sep=';'
)


# =========================================
# 2. Encode categorical variables
# =========================================
def Encoder(df):
    df = df.copy()
    cols = df.select_dtypes(include=['object']).columns
    le = LabelEncoder()
    for col in cols:
        df[col] = le.fit_transform(df[col])
    return df

df = Encoder(df)


# =========================================
# 3. Split features and target
# =========================================
y = df['y']
X = df.drop('y', axis=1)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=17
)


# =========================================
# 4. Default model accuracies (needed for Q2b comparison)
# =========================================
default_models = {
    "Random Forest": RandomForestClassifier(random_state=17),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=17),
    "LightGBM": LGBMClassifier(random_state=17),
    "Gradient Boosting": GradientBoostingClassifier(random_state=17)
}

default_results = {}

print("===== DEFAULT MODEL TEST ACCURACIES =====")
for name, model in default_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    default_results[name] = acc
    print(f"{name}: {acc:.4f}")


# =========================================
# 5. Optuna objective functions
# =========================================

# Random Forest:
# "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1])
# "max_features": trial.suggest_int("max_features", 2, X_train.shape[1])

def objective_rf(trial):
    params = {
        "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1]),
        "max_features": trial.suggest_int("max_features", 2, X_train.shape[1]),
        "random_state": 17,
        "n_jobs": -1
    }
    model = RandomForestClassifier(**params)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy', n_jobs=-1).mean()
    return score


# XGBoost:
# "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1])
# "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.3, log=True)

def objective_xgb(trial):
    params = {
        "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1]),
        "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.3, log=True),
        "random_state": 17,
        "eval_metric": "logloss"
    }
    model = XGBClassifier(**params)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy', n_jobs=-1).mean()
    return score


# LightGBM:
def objective_lgb(trial):
    params = {
        "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1]),
        "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.3, log=True),
        "random_state": 17
    }
    model = LGBMClassifier(**params)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy', n_jobs=-1).mean()
    return score


# Gradient Boosting:
def objective_gb(trial):
    params = {
        "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1]),
        "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.3, log=True),
        "random_state": 17
    }
    model = GradientBoostingClassifier(**params)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy', n_jobs=-1).mean()
    return score


# =========================================
# 6. Run Optuna studies (100 trials each)
# =========================================
print("\n===== OPTUNA TUNING STARTED =====")

study_rf = optuna.create_study(direction='maximize')
study_rf.optimize(objective_rf, n_trials=100)

study_xgb = optuna.create_study(direction='maximize')
study_xgb.optimize(objective_xgb, n_trials=100)

study_lgb = optuna.create_study(direction='maximize')
study_lgb.optimize(objective_lgb, n_trials=100)

study_gb = optuna.create_study(direction='maximize')
study_gb.optimize(objective_gb, n_trials=100)


# =========================================
# 7. Print best parameters and CV scores
# =========================================
print("\n===== BEST PARAMETERS AND CV ACCURACIES =====")
print("Random Forest Best Params:", study_rf.best_params)
print("Random Forest Best CV Accuracy:", round(study_rf.best_value, 4))

print("\nXGBoost Best Params:", study_xgb.best_params)
print("XGBoost Best CV Accuracy:", round(study_xgb.best_value, 4))

print("\nLightGBM Best Params:", study_lgb.best_params)
print("LightGBM Best CV Accuracy:", round(study_lgb.best_value, 4))

print("\nGradient Boosting Best Params:", study_gb.best_params)
print("Gradient Boosting Best CV Accuracy:", round(study_gb.best_value, 4))


# =========================================
# 8. Find overall best method from CV
# =========================================
studies = {
    "Random Forest": study_rf,
    "XGBoost": study_xgb,
    "LightGBM": study_lgb,
    "Gradient Boosting": study_gb
}

best_method = max(studies, key=lambda k: studies[k].best_value)
best_study = studies[best_method]
best_params = best_study.best_params

print("\n===== BEST METHOD FROM CROSS VALIDATION =====")
print("Best Method:", best_method)
print("Best Parameters:", best_params)
print("Best CV Accuracy:", round(best_study.best_value, 4))


# =========================================
# 9. Train tuned best model on training data
#    and evaluate on test data
# =========================================
if best_method == "Random Forest":
    tuned_model = RandomForestClassifier(
        **best_params,
        random_state=17,
        n_jobs=-1
    )

elif best_method == "XGBoost":
    tuned_model = XGBClassifier(
        **best_params,
        random_state=17,
        eval_metric='logloss'
    )

elif best_method == "LightGBM":
    tuned_model = LGBMClassifier(
        **best_params,
        random_state=17
    )

elif best_method == "Gradient Boosting":
    tuned_model = GradientBoostingClassifier(
        **best_params,
        random_state=17
    )

tuned_model.fit(X_train, y_train)
y_pred_tuned = tuned_model.predict(X_test)
tuned_test_acc = accuracy_score(y_test, y_pred_tuned)

print("\n===== TEST PERFORMANCE OF BEST TUNED METHOD =====")
print("Best Tuned Method:", best_method)
print("Tuned Test Accuracy:", round(tuned_test_acc, 4))


# =========================================
# 10. Compare with default version of same method
# =========================================
default_same_method_acc = default_results[best_method]

print("\n===== COMPARISON WITH DEFAULT VERSION =====")
print(f"Default {best_method} Test Accuracy:", round(default_same_method_acc, 4))
print(f"Tuned   {best_method} Test Accuracy:", round(tuned_test_acc, 4))

if tuned_test_acc > default_same_method_acc:
    print("Improvement: Yes")
elif tuned_test_acc < default_same_method_acc:
    print("Improvement: No, tuned model is worse")
else:
    print("Improvement: No change")
```

```python
#QUESTION 4

# =========================================
# QUESTION 4
# Optuna tuning with 100 trials using CV=3
# Compare tuned vs default using R2
# =========================================

# Install if needed:
# !pip install optuna xgboost lightgbm

import pandas as pd
import numpy as np
import optuna

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor


# =========================================
# 1. Load Data
# =========================================
dr = pd.read_csv('https://raw.githubusercontent.com/ogut77/DataScience/main/data/diamond.csv')


# =========================================
# 2. Encode categorical variables
# =========================================
def Encoder(df):
    df = df.copy()
    cols = df.select_dtypes(include=['object']).columns
    le = LabelEncoder()
    for col in cols:
        df[col] = le.fit_transform(df[col])
    return df

dr = Encoder(dr)


# =========================================
# 3. Split features and target
# =========================================
y = dr['Price']
X = dr.drop('Price', axis=1)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=17
)


# =========================================
# 4. Default model R2 scores for comparison
# =========================================
default_models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=17),
    "Random Forest": RandomForestRegressor(random_state=17, n_jobs=-1),
    "XGBoost": XGBRegressor(random_state=17, n_jobs=-1),
    "LightGBM": LGBMRegressor(random_state=17),
    "Gradient Boosting": GradientBoostingRegressor(random_state=17)
}

default_results = {}

print("===== DEFAULT MODEL TEST R2 SCORES =====")
for name, model in default_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    default_results[name] = r2
    print(f"{name}: {r2:.4f}")


# =========================================
# 5. Optuna objective functions
# =========================================

# Random Forest:
# "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1])
# "max_features": trial.suggest_int("max_features", 2, X_train.shape[1])

def objective_rf(trial):
    params = {
        "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1]),
        "max_features": trial.suggest_int("max_features", 2, X_train.shape[1]),
        "random_state": 17,
        "n_jobs": -1
    }
    model = RandomForestRegressor(**params)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='r2', n_jobs=-1).mean()
    return score


# XGBoost:
# "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1])
# "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.3, log=True)

def objective_xgb(trial):
    params = {
        "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1]),
        "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.3, log=True),
        "random_state": 17,
        "n_jobs": -1
    }
    model = XGBRegressor(**params)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='r2', n_jobs=-1).mean()
    return score


# LightGBM
def objective_lgb(trial):
    params = {
        "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1]),
        "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.3, log=True),
        "random_state": 17
    }
    model = LGBMRegressor(**params)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='r2', n_jobs=-1).mean()
    return score


# Gradient Boosting
def objective_gb(trial):
    params = {
        "max_depth": trial.suggest_int("max_depth", 2, X_train.shape[1]),
        "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.3, log=True),
        "random_state": 17
    }
    model = GradientBoostingRegressor(**params)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='r2', n_jobs=-1).mean()
    return score


# =========================================
# 6. Run Optuna studies
# =========================================
print("\n===== OPTUNA TUNING STARTED =====")

study_rf = optuna.create_study(direction='maximize')
study_rf.optimize(objective_rf, n_trials=100)

study_xgb = optuna.create_study(direction='maximize')
study_xgb.optimize(objective_xgb, n_trials=100)

study_lgb = optuna.create_study(direction='maximize')
study_lgb.optimize(objective_lgb, n_trials=100)

study_gb = optuna.create_study(direction='maximize')
study_gb.optimize(objective_gb, n_trials=100)


# =========================================
# 7. Print best parameters and CV R2
# =========================================
print("\n===== BEST PARAMETERS AND CV R2 SCORES =====")
print("Random Forest Best Params:", study_rf.best_params)
print("Random Forest Best CV R2:", round(study_rf.best_value, 4))

print("\nXGBoost Best Params:", study_xgb.best_params)
print("XGBoost Best CV R2:", round(study_xgb.best_value, 4))

print("\nLightGBM Best Params:", study_lgb.best_params)
print("LightGBM Best CV R2:", round(study_lgb.best_value, 4))

print("\nGradient Boosting Best Params:", study_gb.best_params)
print("Gradient Boosting Best CV R2:", round(study_gb.best_value, 4))


# =========================================
# 8. Find overall best tuned method by CV R2
# =========================================
studies = {
    "Random Forest": study_rf,
    "XGBoost": study_xgb,
    "LightGBM": study_lgb,
    "Gradient Boosting": study_gb
}

best_method = max(studies, key=lambda k: studies[k].best_value)
best_study = studies[best_method]
best_params = best_study.best_params

print("\n===== BEST METHOD FROM CROSS VALIDATION =====")
print("Best Method:", best_method)
print("Best Parameters:", best_params)
print("Best CV R2:", round(best_study.best_value, 4))


# =========================================
# 9. Train best tuned method and test it
# =========================================
if best_method == "Random Forest":
    tuned_model = RandomForestRegressor(
        **best_params,
        random_state=17,
        n_jobs=-1
    )

elif best_method == "XGBoost":
    tuned_model = XGBRegressor(
        **best_params,
        random_state=17,
        n_jobs=-1
    )

elif best_method == "LightGBM":
    tuned_model = LGBMRegressor(
        **best_params,
        random_state=17
    )

elif best_method == "Gradient Boosting":
    tuned_model = GradientBoostingRegressor(
        **best_params,
        random_state=17
    )

tuned_model.fit(X_train, y_train)
y_pred_tuned = tuned_model.predict(X_test)
tuned_test_r2 = r2_score(y_test, y_pred_tuned)

print("\n===== TEST PERFORMANCE OF BEST TUNED METHOD =====")
print("Best Tuned Method:", best_method)
print("Tuned Test R2:", round(tuned_test_r2, 4))


# =========================================
# 10. Compare with default version of same method
# =========================================
default_same_method_r2 = default_results[best_method]

print("\n===== COMPARISON WITH DEFAULT VERSION =====")
print(f"Default {best_method} Test R2:", round(default_same_method_r2, 4))
print(f"Tuned   {best_method} Test R2:", round(tuned_test_r2, 4))

if tuned_test_r2 > default_same_method_r2:
    print("Improvement: Yes")
elif tuned_test_r2 < default_same_method_r2:
    print("Improvement: No, tuned model is worse")
else:
    print("Improvement: No change")
```
